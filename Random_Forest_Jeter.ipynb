```python
# import libraries
import pandas as pd
import numpy as np
import os
from sklearn.ensemble import RandomForestClassifier
import datetime
from dateutil.parser import parse
```

```python
# csv file path
all_incidents_csv = "incident_data/yearly_incident_data/all_incidents.csv"

# Import the csv file into data frame
all_incidents_df = pd.read_csv(all_incidents_csv, encoding="utf-8", low_memory = False)



all_incidents_df.head()
# print(all_incidents_df.dtypes)
```

```python
target = all_incidents_df['ucrCode']
target_binary_encoded = pd.get_dummies(target)
target_binary_encoded.head()
# target
```

```python
#Drop columns not needed ('publicAddress', 'caseNumber','offense', 'description', 'ucrCode') 
# These columns did not work in the model so droped them ('lat', 'lon','incidentDate', 'incidentTime' )

data_X = all_incidents_df.drop(['publicAddress', 'caseNumber','offense','description', 'ucrCode', 'lat', 'lon','incidentDate', 'incidentTime'], axis=1)
data_X.head()
```

```python
# This made Precint and neighborhood into binary values for the model

data = data_X.copy()

data_binary_encoded = pd.get_dummies(data)
data_binary_encoded.head()
```

```python
# This split the data into training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_binary_encoded, target_binary_encoded, random_state=42)
# y_train
```

```python
#  This is the Random Forest model.  It has a very low fit score.
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```
